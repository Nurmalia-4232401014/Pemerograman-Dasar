Perintah df.duplicated().sum() digunakan untuk mengecek apakah ada baris data yang duplikat di dalam DataFrame. Fungsi duplicated() menandai baris yang isinya sama persis dengan baris sebelumnya sebagai True, lalu sum() menjumlahkan berapa banyak baris yang terdeteksi sebagai duplikat. Dengan cara ini, kita bisa mengetahui apakah dataset memiliki data ganda yang perlu dibersihkan agar analisis tidak bias.

```python
# Mengisi nilai kosong di kolom numerik dengan nilai rata-rata
df['Umur'] = df['Umur'].fillna(df['Umur'].mean())
df['Nilai_Tugas'] = df['Nilai_Tugas'].fillna(df['Nilai_Tugas'].mean())
df['Nilai_UAS'] = df['Nilai_UAS'].fillna(df['Nilai_UAS'].mean())

# Mengisi nilai kosong di kolom kategorik dengan 'Tidak Diketahui'
df['Jurusan'] = df['Jurusan'].fillna('Tidak Diketahui')

df
```

Program tersebut digunakan untuk membersihkan data yang memiliki nilai kosong. Untuk kolom yang berisi angka seperti Umur, Nilai_Tugas, dan Nilai_UAS, nilai kosong diganti dengan nilai rata-rata dari masing-masing kolom agar data tetap konsisten dan bisa diolah tanpa masalah. Sementara itu, untuk kolom kategori seperti Jurusan, nilai kosong diganti dengan teks Tidak Diketahui sebagai penanda bahwa informasi tersebut tidak tercatat. Setelah proses ini, DataFrame menjadi lebih rapi dan lengkap sehingga siap digunakan untuk analisis lebih lanjut.

```python
# Tambahkan satu data duplikat untuk simulasi
df.loc[len(df)] = ['Aldi', 20, 85, 80, 'Energi']

# Cek jumlah duplikat
print("Jumlah duplikat sebelum dihapus:", df.duplicated().sum())

# Hapus duplikat
df = df.drop_duplicates()

print("Jumlah duplikat setelah dihapus:", df.duplicated().sum())
```

Program ini menambahkan satu baris data yang sama persis dengan data sebelumnya untuk mensimulasikan adanya duplikasi. Setelah itu, program menghitung jumlah baris duplikat yang muncul di DataFrame menggunakan df.duplicated().sum(). Selanjutnya, semua data yang terdeteksi duplikat dihapus menggunakan drop_duplicates(). Terakhir, program kembali mengecek jumlah duplikat untuk memastikan bahwa seluruh data ganda telah berhasil dihapus sehingga dataset menjadi bersih dan tidak memiliki pengulangan data.
